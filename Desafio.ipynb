```python
import numpy as np
temperaturas = np.random.randint(20,36,7)

dias = ["Domingo","Segunda","Terça","Quarta","Quinta","Sexta","Sabado"]

print("Temperaturas:",temperaturas)
print("Dias:",dias)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Temperaturas": temperaturas,
    "Dias": dias
})


media = df["Temperaturas"].mean().round(1)

max = df["Temperaturas"].max()
min = df["Temperaturas"].min()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(df["Dias"], df["Temperaturas"])
plt.title("Temperaturas x Dias da Semana")
plt.xlabel = "Dias da Semana"
plt.ylabel = "Temperaturas"
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(df["Dias"], df["Temperaturas"])
plt.title("Dias da Semana x Temperaturas")
plt.xlabel = "Dias da Semana"
plt.ylabel= "Temperaturas"
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df["Dias"],df["Temperaturas"])
plt.title("Dias da Semana x Temperaturas")
plt.xlabel = "Dias da Semana"
plt.ylabel = "Temperaturas"
plt.show()
```
