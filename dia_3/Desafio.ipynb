```python
import numpy as np #importar a biblioteca numpy para realizar calculos

temperaturas = np.random.randint(20,36,7) #Gerar e armazenar 7 temperaturas aleatorias entre 20 e 35 com a função random.randint

dias = ["Domingo","Segunda","Terça","Quarta","Quinta","Sexta","Sabado"] #Criar uma lista com os dias da semana
```

```python
import pandas as pd #importar a biblioteca pandas para organizar e tratar os dados

df = pd.DataFrame({  #Criar um dataframe com as temperaturas geradas e seus respectivos dias da semana
    "Temperaturas": temperaturas,
    "Dias": dias
})


media = df["Temperaturas"].mean().round(1) #Armazenar a media das temperaturas pela função .mean() e arredondar para uma casa decimal com a função round(1)

max = df["Temperaturas"].max() #Armazenar a temperatura maxima pela função .max()
min = df["Temperaturas"].min() #Armazenar a temperatura minima pela função .min()

```

```python
import matplotlib.pyplot as plt #importar a biblioteca plot para criar os graficos e suas representações
#Criar um grafico de barras pela biblioteca plot. Com o eixo x representando os dias da semana e o eixo y as temperaturas
plt.figure()
plt.bar(df["Dias"], df["Temperaturas"])
plt.title("Temperaturas x Dias da Semana")
plt.xlabel = "Dias da Semana"
plt.ylabel = "Temperaturas"
plt.show()
#Esses graficos ajudam na visualização das variações da amplitude térmica, observando os picos de alta e baixa temperaturas entre os dias da semana
```

```python
import matplotlib.pyplot as plt #importar a biblioteca plot para criar os graficos e suas representações
#Criar um grafico de linhas pela biblioteca plot. Com o eixo x representando os dias da semana e o eixo y as temperaturas
plt.figure()
plt.plot(df["Dias"], df["Temperaturas"])
plt.title("Dias da Semana x Temperaturas")
plt.xlabel = "Dias da Semana"
plt.ylabel= "Temperaturas"
plt.show()
#Esses graficos ajudam na visualização das variações da amplitude térmica, observando os picos de alta e baixa temperaturas entre os dias da semana
```

```python
import matplotlib.pyplot as plt #importar a biblioteca plot para criar os graficos e suas representações
#Criar um grafico de dispersão pela biblioteca plot. Com o eixo x representando os dias da semana e o eixo y as temperaturas
plt.figure()
plt.scatter(df["Dias"],df["Temperaturas"])
plt.title("Dias da Semana x Temperaturas")
plt.xlabel = "Dias da Semana"
plt.ylabel = "Temperaturas"
plt.show()
#Esses graficos ajudam na visualização das variações da amplitude térmica, observando os picos de alta e baixa temperaturas entre os dias da semana
```
